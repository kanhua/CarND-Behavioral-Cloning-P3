```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import cv2
```

```python
def clahe_image(image):
    yuv_img = cv2.cvtColor(image,cv2.COLOR_RGB2HSV)
    clahe = cv2.createCLAHE(clipLimit=5.0, tileGridSize=(4,4))
    yuv_img[:,:,2]= clahe.apply(yuv_img[:,:,2])
    n_img = cv2.cvtColor(yuv_img,cv2.COLOR_HSV2RGB)
    return n_img
```

```python
sample_image_file='./examples/uphill.jpg'
temp_img=cv2.imread(sample_image_file)
temp_img=temp_img.astype('uint8')
plt.imshow(temp_img)
plt.show()
```

```python
n_img=clahe_image(temp_img)
```

```python
plt.imshow(n_img)
plt.show()
```

```python
cv2.imwrite("./examples/adj_death_corner.jpg",n_img)
```
